**Conceptos básicos de la estadistica descriptiva**

**Medidas**

**1. media aritmética:** promedio

**2. desviación estándar:**depende de ls medis de mis datos

**3. varianza:**disperción (que tan dispersos están los datos/medidas) variable unidemensional

**4. desviación típica**:con respecto a la varianza

**5. moda:** valor/dato que aparece con más frecuencia (se repite)

**6. mediana:** el lugar/valor central de los datos, cuando los datos están ordenados de mayor a menor

**7. correlación:**una relación/dependencia de las variables, deben ser módelos bidemensionales

**8. covarianza:**variables bidemensionales, aplica como se compartan las variables

**9. valor atípico:**valores lejanos a una tendencia/moda, halla problemas, el numero de datos atípicos debe ser menor, si se hallan se deben eliminar ya que entorpecen la información de los datos

Cuando una distribución de frecuencia es simétrica, la
media, mediana y moda coinciden en su valor ( X = Me
= Mo). En el caso de una distribución binomial simétrica,
es necesario calcular el promedio de las modas.

En una distribución sesgada a la izquierda, la moda es
menor a la mediana, y esta a su vez menor que la
media.
* 5. (Mo > Me > X)

En una distribución sesgada a la derecha la relación
se invierte, la moda es mayor a la mediana, y esta a
su vez mayor que la media.

```python
# Ejemplo de diagrama de cajas
datos1 =np.random.normal(100,10,200)
datos3 =np.random.normal(80,30,200)
datos4 =np.random.normal(90,20,200)
datos5 =np.random.normal(70,25,200)
# lista convertilos, se crea una nueva lista donde se concatena (una) las anteriores listas de datos que abriamos creeado antes
datos_graf= [datos1,datos3,datos4,datos5]
#crear la grafica
fig = plt.figure(1, figsize=(9,6)) #el 1  indica el diagrama de bigotes o de caja
ax = fig.add_subplot(111)

ax.boxplot(datos_graf)
 #flier marcadores valores atipicos
# los puntos fuera son los datos atipicos
```

```python
#otra manera
fig = plt.figure(1, figsize=(9,6)) #el 1  indica el diagrama de bigotes o de caja
ax = fig.add_subplot(111)

bp=ax.boxplot(datos_graf)

for flier in bp ["fliers"]:
  flier.set(marker="o", markerfacecolor="orange") #flier permite dar la funcion de crear marcador su forma: marker"o" y su color markerfacecolor:"red". y en este caso marcamos los valores atipicos
# los puntos fuera son los datos atipicos
```

** REGRESIÓN LINEAL CON PHYTON Y SKLEARN **

creamos nuestros datos de entrada, por el momento solo Word count y como etiquetas los # shares. creamos el objeto linear regression y lo hacemos "encajar"(entrenar) con el metodo fit(). Finalmente imprimimos los coeficientes y puntajes obtenidos

```python
## ASIGNAMOS NUESTRA VARIABLE DE ENTRADA X PARA ENTRENAMIENTO Y LAS ETIQUETAS Y
dataX = filtered_data[["Word count"]]
x_train = np.array(dataX)
y_train = filtered_data["# Shares"].values

## CREAMOS EL IBJETO DE REGRESION LINEAR
regr = linear_model.LinearRegression()

##ENTRENAMOS NUESTRO MODELO
regr.fit(x_train, y_train)

##HACEMOS LAS PREDICCIONES QUE EN DEFINITIVA UNA LINEA (EN ESTE CASO AL SER 20)
y_pred = regr.predict(x_train)


##VEREMOS LOS COEFICIENTES OBTENIDOS EN NUESTRO CASO, SERÁ LA TANGENTE
print("Coefficients: \n', regr.coef_")

##ESTE ES EL VALOR DONDE CORTA EL EJE Y (EN X=0)
print("Independent term: \n', regr.intercept_ ")

##ERROR AL CUADRADO
print("Mean squared error: %.2f" % mean_squared_error(y_train, y_pred))

##PUNTAJE DE VARIANZA EL MEJOR PUNTAJE SERIA 1.0
print("Variance score: %.2f" % r2_score(y_train, y_pred))
```

De la ecuacion de la recta y = mX + b la pendiente "m" es el coeficiente 5.69 y el término independiente de "b" es 11200. hay un error cuadrático medio enorme por lo que en realidad este modelo no seria bueno, se ve támbien reflejado en el puntaje de varianza que deberia ser cercano a 1.0
